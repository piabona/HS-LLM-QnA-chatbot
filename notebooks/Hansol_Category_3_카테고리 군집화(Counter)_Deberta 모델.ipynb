```python
import pandas as pd
import ast # Abstract Syntax Trees(추상 구문 트리)

kiwi_train['kiwi_questions'] = kiwi_train['kiwi_questions'].apply(ast.literal_eval)
kiwi_train['kiwi_answers'] = kiwi_train['kiwi_answers'].apply(ast.literal_eval)
```

```python
kiwi_train

kiwi_question = kiwi_train['kiwi_questions'].tolist()
kiwi_answer = kiwi_train['kiwi_answers'].tolist()
```

```python
kiwi_question_vocabulary = []

for i in tqdm(kiwi_question) :
    for j in i :
        kiwi_question_vocabulary.append(j)
```

```python
from collections import Counter
kiwi_question_vocabulary_tuple = tuple(kiwi_question_vocabulary)

Counter(kiwi_question_vocabulary_tuple).most_common()
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")

sequence_to_classify = "Angela Merkel ist eine Politikerin in Deutschland und Vorsitzende der CDU"
candidate_labels = ["politics", "economy", "entertainment", "environment"]
output = classifier(sequence_to_classify, candidate_labels, multi_label=False)
print(output)
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")

sequence_to_classify = "면진장치가 뭐야?"
candidate_labels_1 = category.copy()
candidate_labels_2 = qna_cat.copy()
output_1 = classifier(sequence_to_classify, candidate_labels_1, multi_label=True)
output_2 = classifier(sequence_to_classify, candidate_labels_2, multi_label=True)

print(output_1)
print(output_2)
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")

sequence_to_classify = "내진설계의 종류 좀 알려줘"
candidate_labels_1 = category.copy()
candidate_labels_2 = qna_cat.copy()
output_1 = classifier(sequence_to_classify, candidate_labels_1, multi_label=True)
output_2 = classifier(sequence_to_classify, candidate_labels_2, multi_label=True)

print(output_1)
print(output_2)
```
