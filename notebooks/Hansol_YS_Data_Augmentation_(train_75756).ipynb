## 2-1. 동일 카테고리 (80%)

```python
def idcat_random_match_select(data, category, total_n=1000):
    tmp = data[data['category'] == category]
    tmp = tmp.reset_index().drop('index', axis=1)

    # 카테고리별 % 반영
    cat = gpt_df['category'].unique().tolist()
    k = int(pcs[cat.index(category)]/100 * total_n)

    que_list = []
    ans_list = []

    for _ in tqdm(range(k)):
        # 연결어
        connect_words = ['그리고', '또', '또한', '혹은', '이와 함께', '게다가', '그뿐 아니라', '한편']
        selected_connect = random.choice(connect_words)

        # 콤마
        comma_words = ['', ',']
        selected_comma = random.choice(comma_words)

        # index 조합
        a,b = random.sample(range(len(tmp)), 2)
        que = tmp['question'][a] +' '+ selected_connect + selected_comma +' '+ tmp['question'][b]
        ans = tmp['answer'][a] +' '+ selected_connect + selected_comma +' '+ tmp['answer'][b]
        que_list.append(que)
        ans_list.append(ans)

    return que_list, ans_list, category
```

```python
def diffcat_random_match_select(data, total_n):

    cat = data['category'].unique().tolist()
    que_list = []
    ans_list = []
    cat_list = []
    for _ in tqdm(range(total_n)):

        # 연결어
        connect_words = ['그리고', '또', '또한', '혹은', '이와 함께', '게다가', '그뿐 아니라', '한편']
        selected_connect = random.choice(connect_words)

        # 콤마
        comma_words = ['', ',']
        selected_comma = random.choice(comma_words)

        # index 조합
        cat1,cat2 = random.sample(cat, 2)

        tmp1 = data[data['category'] == cat1].reset_index(drop=True)
        tmp2 = data[data['category'] == cat2].reset_index(drop=True)

        a = random.sample(range(len(tmp1)), 1)[0]
        b = random.sample(range(len(tmp2)), 1)[0]

        que = tmp1['question'][a] +' '+ selected_connect + selected_comma +' '+ tmp2['question'][b]
        ans = tmp1['answer'][a] +' '+ selected_connect + selected_comma +' '+ tmp2['answer'][b]
        new_cat = [cat1, cat2]

        que_list.append(que)
        ans_list.append(ans)
        cat_list.append(new_cat)

    return que_list, ans_list, cat_list
```

```python
def diffcat_make_df(data, total_n) :
    tmp = pd.DataFrame(columns=['question','answer', 'category'])
    que_list, ans_list, cat_list = diffcat_random_match_select(df, total_n)
    tmp = pd.concat([tmp, pd.DataFrame({'question':que_list, 'answer':ans_list, 'category':cat_list})], axis=0)
    return tmp
```

# 데이터셋

```python
# dataset 클래스 생성

class ChatDataset(torch.utils.data.Dataset):
    def __init__(self,df):
        self.df = df
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx) :
        item = self.df.iloc[idx]
        question = item['question']
        answer = item['answer']

        return {
            'question': question,
            'answer': answer,
            'category' : item['category']
        }
```

```python
def collate_fn(batch):
    texts = [f"<q>{item['question']}</s><a>{item['answer']}</s>{item['category']}"
    for item in batch]

    x = tokenizer(texts, return_tensors='pt', padding=True)
    return {'x':x}
```

```python
# reset_seeds(SEED)

from tqdm import tqdm

def train_loop(dataloader, model, loss_fn, optimizer, device):
    epoch_loss = 0
    model.train()
    for batch in tqdm(dataloader):
        x = batch['x'].to(device) # 학습을 위해서 device로 옮김

        pred = model(**x).logits # batch, seq, 단어별
        n_class = pred.shape[-1] # kogpt2 출력수 51200

        pred = pred[:, :-1, :] # 마지막 차원의 마지막 토큰인 eos 토큰만 제외함
        pred = pred.reshape(-1, n_class) # 2차원 형태로 변환

        tgt = x['input_ids'][:,1:]
        tgt = tgt.flatten() # 1차원 벡터 형태로 바뀜

        mask = tgt != tokenizer.pad_token_id # 손실 계산시 제외
        tgt = tgt[mask]
        pred = pred[mask]
        loss = loss_fn(pred, tgt)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    epoch_loss /= len(dataloader)
    return epoch_loss
```

```python
from datetime import date
today = date.today()

today_exp = f'{today.month}_{today.day}'
today_exp
```

```python

reset_seeds(SEED)

batch_size = 2
epochs = 10

from transformers import AutoTokenizer, AutoModelForCausalLM
model_name = "skt/kogpt2-base-v2"
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=3e-5)
loss_fn = torch.nn.CrossEntropyLoss()

train_dt = ChatDataset(train_ft)
train_dl = torch.utils.data.DataLoader(train_dt,batch_size=batch_size,shuffle=True,collate_fn=collate_fn)

for i in range(epochs):
    train_loss = train_loop(train_dl, model, loss_fn, optimizer, device)
    print(f"{i+1}번째 epoch: {train_loss:.4f}")
    model.save_pretrained(f"{data_path}models/kogpt2_trial4_date_{today_exp}_epoch{i+1}_loss{train_loss:.4f}")
```
